#### 1️⃣ 5 Analysis Questions for Your Project

I have use these questions in my analysis.

1️ Which product generates the highest total sales?

2️ Which city has the highest revenue?

3️ Which category sells the most products (Electronics vs Accessories)?

4️ Which month has the highest sales?

5️ Top 5 best selling products based on quantity.

```python
## Q-1:Which product generates the highest total sales?
highest_total_sales=df.groupby(["Product"],as_index=False)["Total_Sales"].sum().sort_values(by=["Total_Sales"],ascending=False)
```

```python
highest_total_sales
```

#### Visualize the highest_total_sales:

```python
ax=sns.barplot(x="Product",y="Total_Sales",data=highest_total_sales)
for bar in ax.containers:
    ax.bar_label(bar,fmt="%.0f",fontsize=8)

plt.title("highest_total_sales")
plt.xticks(rotation=45)
plt.xlabel("Product")
plt.ylabel("Total_Sales")
plt.show()
```

```python
## Q-2:Which city has the highest revenue?
highest_revenue=df.groupby(["City"],as_index=False)["Total_Sales"].sum().sort_values(by=["Total_Sales"],ascending=False)
print(highest_revenue)
```

#### Visualize the highest_revenue:

```python
bx=sns.barplot(x="City",y="Total_Sales",data=highest_revenue)
for bar in bx.containers:
    bx.bar_label(bar,fmt="%.0f")

plt.title("city wise the highest revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
## Which category sells the most products (Electronics vs Accessories)?
category_sales = df.groupby("Category",as_index=False)["Quantity"].sum().sort_values(by=["Quantity"],ascending=False)

top_category = category_sales.idxmax()

print("Category with most products sold:", top_category)
print(category_sales)
```

#### Visualization of Category & Quantity:


```python
cx=sns.barplot(x="Category",y="Quantity",data=category_sales)
for bar in cx.containers:
    cx.bar_label(bar,fmt="%.0f")


plt.title("Category wise Quantity")
plt.show()
```

#### Q-3:Which month has the highest sales?

```python
df["Date"] = pd.to_datetime(df["Date"])

df["Month"] = df["Date"].dt.month

monthly_sales = df.groupby("Month")["Total_Sales"].sum()

top_month = monthly_sales.idxmax()

print("Month with highest sales:", top_month)
print(monthly_sales)

#Convert to DataFrame for seaborn
monthly_sales = monthly_sales.reset_index()
```

#### #### visualization of line chart (numerical + numerical data so that's why I use line chart)

```python
# Line plot
sns.lineplot(data=monthly_sales, x="Month", y="Total_Sales", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.show()
```

#### Top 5 best selling products based on quantity

```python
top_products = df.groupby(["Product"],as_index=False)["Quantity"].sum().sort_values(by=["Quantity"],ascending=False).head(5)

print("Top 5 best selling products:")
print(top_products)
```

#### Visualization the top 5 quantity wise products

```python
dx=sns.barplot(x="Product",y="Quantity",data=top_products)
for bar in dx.containers:
    dx.bar_label(bar,fmt="%.0f")
plt.title("Quantity wise saling top 5 products")
plt.show()
```
